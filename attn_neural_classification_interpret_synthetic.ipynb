
Assume that we have a vacobulary of tokens labeled as 1, 2, ..., N. There are synthetic sentences consisting of the tokens, each sentence has the same length. Assume we have some three types of tokens: positive, negative and neutral, Positive (negative) tokens only appear in positive (negative) sentences, neutral tokens appear evenly across all the sentences.

```python
import numpy as np
import torch 
import copy
from torch import nn
from torch.nn import functional as F
from matplotlib import pyplot as plt
%matplotlib inline
#plt.style.use('ggplot')
np.random.seed(2)
```

### Create Vocabulary

```python
#tokens in order
vocab = list(range(1000))
pos_words = range(1000, 1050)
neg_words = range(1050, 1100)
neutral_words = range(1000)

# #tokens at random
# vocab = list(range(1100))
# senti_words = np.random.choice(len(vocab), 100, replace=False)
# pos_words = senti_words[:50]
# neg_words = senti_words[50:]
# neutral_words = [w for w in vocab if w not in senti_words]

#overlap_words = range(1100, 1150)

sent_len = 10
pos_sent_num = 6000
neg_sent_num = 6000
```

### Generate Sentences

```python
#Generate sentences
pos_sents = []
neg_sents = []
rate = 8000
for i in range(pos_sent_num):
  sent = np.random.choice(neutral_words, sent_len, replace=False)
  sent = list(sent)
  if i<rate:
    sent += list(np.random.choice(pos_words, 2, replace=False))
  else:
    sent += list(np.random.choice(neg_words, 2, replace=False))
  pos_sents.append(sent)

  
#let some positive words appear in negative sentences
for i in range(neg_sent_num):
  sent = np.random.choice(neutral_words, sent_len, replace=False)
  sent = list(sent)
  if i<rate:
    sent += list(np.random.choice(neg_words, 2, replace=False))
  else:
    sent += list(np.random.choice(pos_words, 2, replace=False))
  neg_sents.append(sent)
```

```python
total_sents = pos_sents + neg_sents
labels = [1]*pos_sent_num+[0]*neg_sent_num
train_data = list(zip(total_sents, labels))
np.random.shuffle(train_data)
```

### Model with Scaled Dot-product Attention

```python
from torch.nn import utils as nn_utils
import torch
import pickle
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import Variable
import torch.nn.init as init

scale = 10#10#50#1#5
class AffineAttnClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, label_dim=1, attn_type='dot_prod', use_affine=False):
        super(AffineAttnClassifier, self).__init__()
        
        self.embeddings = nn.Embedding(vocab_size, embed_dim)
        #smaller embeddings may be better
        self.embeddings.weight.data.uniform_(-0.1,0.1)
        self.dropout = nn.Dropout(0.5)

        self.affine = nn.Linear(embed_dim, hidden_dim)
        self.softmax = nn.Softmax(dim=-1)
        self.sigmoid = nn.Sigmoid()
        

        self.attn_linear = nn.Linear(hidden_dim, hidden_dim, bias=False)
        self.V = nn.Parameter(torch.randn(hidden_dim, 1))

        self.decoder = nn.Linear(hidden_dim, label_dim, bias=False)
        self.decoder.weight.data.uniform_(-1, 1)

        self.log_softmax = nn.LogSoftmax(dim=-1)
        
        self.attn_type = attn_type
        self.use_affine = use_affine
        
    def forward(self, seq_ids, seq_lengths=None):
        '''
        Args:
            seq_ids: word indexes, batch_size, max_len, Long Tensor
            seq_lengths: lengths of sentences, batch_size, Long Tensor
        attention:
            score = v h
            att = softmax(score)
        '''
        #For debugging
        #self.embeddings.weight.retain_grad()
        #self.decoder.weight.retain_grad()
        #self.V.retain_grad()
        
        seq_embs = self.embeddings(seq_ids)
        seq_embs = self.dropout(seq_embs)
        batch_size, max_len, hidden_dim = seq_embs.size()
        
        # batch * max_len * hidden_states
        if self.use_affine:
            hidden_vecs = self.affine(seq_embs)
        else:
            hidden_vecs = seq_embs
        #batch * max_len * hidden_states
        if self.attn_type == 'dot_prod':
            inter_out = hidden_vecs
        else:
            inter_out = torch.tanh(self.attn_linear(hidden_vecs))
        
        #batch * max_len
        scores = torch.matmul(inter_out, self.V).squeeze(-1)
        scores = scores/scale
        #Mask the padding values
        mask = torch.zeros(batch_size, max_len)
        for i in range(batch_size):
            mask[i, seq_lengths[i]:] = 1
        scores = scores.masked_fill(mask.bool(), -np.inf)
        #Softmax, batch_size*1*max_len
        attn = self.softmax(scores).unsqueeze(1)
        #weighted sum, batch_size*hidden_dim. Sentence representations
        final_vec = torch.bmm(attn, hidden_vecs).squeeze(1)
        final_vec = self.dropout(final_vec)
        #instance-level polarity score
        outputs = self.decoder(final_vec)
        probs = self.sigmoid(outputs)
        return probs, outputs, attn, scores
```


Note:
- SGD converges slowly and the trend is similar to what we analyse in our paper. There can be a gap between the attention scores of positive tokens and that of negative tokens, this is due to $V^TW$.
- The weight_decay will influence the trends and patterns.
- After sufficient iterations, the inner products between embeddings of the positive/negative tokens are positively large; the inner products between embeddings of the positive and negative tokens are negatively large; the inner products between embeddings of the positive/negative  and neutral tokens are small.
- We have used Dropout to reduce overfitting somehow.
- Adagrad can converge very fast, but the term $V^TW$ can be large.

```python
model = AffineAttnClassifier(1000+100, 100, 100)
#optimizer = optim.Adam(model.parameters(), lr=0.001)
#optimizer = optim.Adagrad(model.parameters(), lr=0.005, lr_decay=0.00)
optimizer = optim.SGD(model.parameters(), lr=0.015, weight_decay=0.000)
loss_function = nn.BCELoss()
```

```python
epochs = 500
batch_size = 200
loops = (pos_sent_num+neg_sent_num)//batch_size
model.train()
#record the trend of attention scores
intermediate_neutral_attn_scores = []
intermediate_pos_attn_scores = []
intermediate_neg_attn_scores = []
intermediate_noisy_attn_scores = []
#record the trend of polarity scores
intermediate_neutral_polarity_scores = []
intermediate_pos_polarity_scores = []
intermediate_neg_polarity_scores = []
#Record the trend of dot-products between word embeddings
intermediate_emb_dot1 = []
intermediate_emb_dot2 = []
intermediate_emb_dot3 = []
intermediate_emb_dot4 = []
intermediate_emb_dot5 = []
intermediate_emb_dot6 = []
model_decoder_weight_list = []
#Record the beta
betas = []
for e in range(epochs):
    #Record the trend of embedding dot-products, tokens randomly selected
    embs = model.embeddings.weight.data[1001:1002]
    embs2 = model.embeddings.weight.data[1004:1005]
    intermediate_emb_dot1.append(torch.matmul(embs, embs2.transpose(0, 1)))
    
    embs = model.embeddings.weight.data[1001:1002]
    embs2 = model.embeddings.weight.data[1053:1054]
    intermediate_emb_dot2.append(torch.matmul(embs, embs2.transpose(0, 1)))
    
    embs = model.embeddings.weight.data[1053:1054]
    embs2 = model.embeddings.weight.data[1073:1074]
    intermediate_emb_dot3.append(torch.matmul(embs, embs2.transpose(0, 1)))
    
    embs = model.embeddings.weight.data[1:2]
    embs2 = model.embeddings.weight.data[8:9]
    intermediate_emb_dot4.append(torch.matmul(embs, embs2.transpose(0, 1)))
    
    embs = model.embeddings.weight.data[1:2]
    embs2 = model.embeddings.weight.data[1003:1004]
    intermediate_emb_dot5.append(torch.matmul(embs, embs2.transpose(0, 1)))
    
    embs = model.embeddings.weight.data[1:2]
    embs2 = model.embeddings.weight.data[1056:1057]
    intermediate_emb_dot6.append(torch.matmul(embs, embs2.transpose(0, 1)))
    #it is better to shuffle the data
    np.random.shuffle(train_data)
    losses = []
    beta_mean = []

    for i in range(loops):
        
        
        start = i*batch_size
        end = (i+1)*batch_size
        batch_data = train_data[start:end]
        texts, labels = zip(*batch_data)
        texts = torch.LongTensor(texts)
        labels = torch.LongTensor(labels)
        seq_lens = torch.LongTensor([12]*batch_size)



        model.zero_grad()
        probs, outputs, attn, _ = model(texts, seq_lens)
        #beta
        beta = torch.sigmoid(-outputs[:, 0]*(labels.float()-0.5)*2)
        beta_mean.append(beta.mean().item())
        

        
        losses.append(probs)
        #outputs = torch.log(scores)
        #loss = loss_function(outputs, labels)
        temp = model.decoder(model.V.transpose(0, 1))
        loss = loss_function(probs.squeeze(), labels.float())# + 0.001*temp[0, 0].abs()
        loss.backward()
        optimizer.step()
        
    if e%10 == 0:
        print(loss.item())
    #Check whether there exists some trend for the metric hV
    betas.append(np.mean(beta_mean))
    attn_score1 = 0
    polarity_score1 = 0
    for w in neutral_words:
        h1 = model.embeddings(torch.LongTensor([w]))
        attn_score1 += torch.matmul(h1, model.V).item()
        polarity_score1 += model.decoder(h1)

    attn_score2 = 0
    polarity_score2 = 0
    for w in pos_words:  
        h2 = model.embeddings(torch.LongTensor([w]))
        attn_score2 += torch.matmul(h2, model.V).item()
        polarity_score2 += model.decoder(h2)

    attn_score3 = 0
    polarity_score3 = 0
    for w in neg_words:
        h3 = model.embeddings(torch.LongTensor([w]))
        attn_score3 += torch.matmul(h3, model.V).item()
        polarity_score3 += model.decoder(h3)



    intermediate_neutral_attn_scores.append(attn_score1/1000)
    intermediate_pos_attn_scores.append(attn_score2/50)
    intermediate_neg_attn_scores.append(attn_score3/50)
    
    intermediate_neutral_polarity_scores.append(polarity_score1/1000)
    intermediate_pos_polarity_scores.append(polarity_score2/50)
    intermediate_neg_polarity_scores.append(polarity_score3/50)

    model_decoder_weight_list.append(copy.deepcopy(model.decoder.weight.data))
    #w*v
    #temp = torch.matmul(model.decoder.weight, model.V)
    #print('W*V:', temp.item())
    #print('Weight norm:', model.decoder.weight.norm(2))
```

**If the $L_2$-norm of the linear layer weight is much smaller than that of the embeddings, it will be likely the dot-products of the embeddings of two positive tokens grow very slow.**

**If the embeddings are initialized with large l2-norms, the model will not be easily to converge.** 

**The $L_2$-norm of $W$ will be increasing when the model is converging**. Before the convergency, the $L_2$-norm may be decreasing first, then increase.

**Dropout can influence the tendancy of the $||W||^2_2$**, without a dropout the $L_2$_norm will increase when the model is converging.

## Attention score
$$a_i = h_iV/\lambda$$
$$\alpha_i=softmax(a_i)$$

We'll look into $a_i$.
